##### 데이터 전처리

```python
#시계열 데이터로 변환

df['계약년월']=pd.to_datetime(df['계약년월'], format='%Y%m')
df['Year'] = df['계약년월'].dt.year
df['Year'] = df['Year'].astype(str)
```

```python
#제곱미터당 거래가로 변환

df['거래금액(만원)'] = df['거래금액(만원)'].str.replace(",", "")
df['거래금액(만원)'] = df['거래금액(만원)'].astype(float)
df['제곱미터당가격(만원)']=df['거래금액(만원)']/df['전용면적(㎡)']
df
```

```python
#부동산 지수 처리
dfR=dfB[dfB['지역']=='대구'].set_index("지역").transpose()
dfR=dfR.reset_index()
dfR.rename(columns = {'index':'날짜'}, inplace = True)
dfR.rename(columns = {'대구':'지수'}, inplace = True)
dfR['날짜']=pd.to_datetime(dfR['날짜'], format='%Y-%m')
dfR
```

```python
dfR[dfR['날짜']=='2021-06-01']['지수']
```

```python
#기존 df에 부동산지수 추가
df = pd.merge(df, dfR, left_on='계약년월', right_on='날짜')
df
```

```python
#행정구 단위 추가
def Sgg(row):
  r = row.split(' ')
  return r[0]+r[1]

df['행정구']=df['시군구'].astype(str).apply(Sgg)
```

```python
dff = {}
dffmean = {}

groups = df.groupby(df['시군구'])
arrstr = df['시군구'].unique()
for st in arrstr:
  tmp=groups.get_group(st)
  dff[st] = tmp
  groups2 = tmp.groupby(tmp['계약년월'])
  arrstr2 = tmp['계약년월'].unique()
  for st2 in arrstr2:
    tmp2=groups2.get_group(st2)
    dffmean[st]=tmp['거래금액(만원)'].sum()/tmp['전용면적(㎡)'].sum()

#dff: 시군구별 자료
#dffmean: 시군구별 전체평균가
```

```python
from datetime import datetime, timedelta

#기준함수선언
#date는 2021-06-01와 같은 식으로 입력
def baseDate(df, date):
  try:
    groups = df.groupby(df['계약년월'])
    tmp = groups.get_group(date)
    rate = tmp['제곱미터당가격(만원)'].mean()
    return rate
  except KeyError:
    date, a, b = date.split('-', 3)
    groups = df.groupby(df['Year'])
    tmp = groups.get_group(date)
    rate = tmp['제곱미터당가격(만원)'].mean()
    return rate

def baseDateAlt(df, date):
  date, a, b = date.split('-', 3)
  groups = df.groupby(df['Year'])
  tmp = groups.get_group(date)
  rate = tmp['제곱미터당가격(만원)'].mean()
  return rate
```
