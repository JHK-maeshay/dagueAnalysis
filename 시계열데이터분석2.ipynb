```python
#plt 임포트, 한글폰트 설치후 런타임 재시작 필요

import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='NanumBarunGothic') 
plt.rcParams['figure.figsize'] = [10, 8] # setting figure size
```

```python
plt.xlabel('2006년')
plt.ylabel('제곱미터당 거래가(만원)')

ax1=plt.subplot(1, 2, 1)
plt.title('황금동')
plt.plot(df1['계약년월'], df1['제곱미터당가격(만원)'], 'o')

ax2=plt.subplot(1, 2, 2, sharey=ax1)
plt.title('산격동')
plt.plot(df2['계약년월'], df2['제곱미터당가격(만원)'], 'o')
```

```python
grouped = df1.groupby(df1['계약년월'])

st1 = []
st2 = []
month = []

for i in range(0,12): 
  st1.append(i+1)

for i in range(0,10): 
  st2.append(i+2010)

for i in range(0,10):
  for j in range(0,12):
    if j<9:
      if df1.get(str(st2[i])+'0'+str(st1[j])) != None:
        datatmp = grouped.get_group(str(st2[i])+'0'+str(st1[j]))
        month[i*12+j] = datatmp['제곱미터당가격(만원)']
    else:
      if df1.get(str(st2[i])+str(st1[j])) != None:
        datatmp = grouped.get_group(str(st2[i])+str(st1[j]))
        month[i*12+j] = datatmp['제곱미터당가격(만원)']

plt.boxplot(month)
```

```python
plt.xlabel('2006년')
plt.ylabel('제곱미터당 거래가(만원)')

ax1=plt.subplot(1, 2, 1)
plt.title('황금동')
plt.boxplot(month)

ax2=plt.subplot(1, 2, 2, sharey=ax1)
plt.title('산격동')
plt.boxplot(month2)
```
